```python
df.drop_duplicates(inplace=True)
df.shape
```

```python
df['rate'].unique()
```

```python
import numpy as np
def CleanRate(value):
    if (value == 'NEW' and value == '-'):
        return np.nan
    else:
        value = str(value).split('/')
        if value[0].replace('.','',1).isdigit():
            return float(value[0])
        else:
            return np.nan

df['rate']=df['rate'].apply(CleanRate)
df['rate'].head(10)
```

```python
df.rate.isnull().sum()
```

```python
df['rate'].fillna(df['rate'].mean(), inplace=True)
df['rate'].isnull().sum()
```

```python
df.info()
```

```python
df.dropna(inplace=True)
df.head(10)
```

```python
df.shape
```

```python
df.rename(columns={'approx_cost(for two people)':'CostFor2People', 'listed_in(type)' : 'TypeOfFood', 'listed_in(city)':'City'}, inplace=True)
df.head()
```

```python
df.drop('City',axis=1, inplace=True)
df.head(10)
```

```python
df['CostFor2People'].unique()
```

```python
def RemoveComma(value):
    value=str(value)
    if ',' in value:
        value=value.replace(',','')
        return float(value)
    else:
        return float(value)
    
df['CostFor2People']=df['CostFor2People'].apply(RemoveComma)
df['CostFor2People'].unique()
```

```python
df.head()
```

```python
rest_type=df['rest_type'].value_counts(ascending=False)
rest_type
```

```python
rest_type_lessThan1000=rest_type[rest_type<1000]
rest_type
```

```python
def Others(value):
    if value in rest_type_lessThan1000:
        return 'Others'
    else:
        return value

df['rest_type']=df['rest_type'].apply(Others)
df['rest_type'].value_counts()
```

```python
df.head(10)
```

```python
location=df['location'].value_counts()
location
```

```python
location_LessThan500 = location[location<500]

def LessLocation(value):
    if value in location_LessThan500:
        return "Others"
    else:
        return value
    
df['location']=df['location'].apply(LessLocation)
df['location'].value_counts()
```
